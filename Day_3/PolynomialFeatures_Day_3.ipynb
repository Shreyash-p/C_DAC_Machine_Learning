```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
```

```python
df = pd.read_csv(r"C:\Users\patil\Desktop\Know IT CADC\Practical Machine Learning\Common_Folder\Day 3\Position_Salaries.csv")
df
```

```python
x = df.iloc[:, 1:-1].values
y = df.iloc[:, -1].values
```

```python
lin_reg = LinearRegression()
lin_reg.fit(x, y)
```

```python
poly_reg = PolynomialFeatures(degree = 4)
```

```python
x_poly = poly_reg.fit_transform(x)
lin_reg_2 = LinearRegression()
lin_reg_2.fit(x_poly, y)
```

```python
#y_pred=lin_reg.predict(X)
plt.scatter(x, y, color = 'red')

plt.plot(x, lin_reg.predict(x), color = 'blue')
plt.title('Predict the salary (Linear Regression)')
plt.xlabel('Position Level')
plt.ylabel('Salary')
plt.show()
```

```python
y_pred_poly=poly_reg.fit_transform(x)
plt.scatter(x, y, color = 'red')
plt.plot(x, lin_reg_2.predict(poly_reg.fit_transform(x)), color = 'blue')
plt.title('Predict the salary(Polynomial Regression)')
plt.xlabel('Position level')
plt.ylabel('Salary')
plt.show()
```
