```python
year = [2001,2002,2003,2004,2005,2006,2007,2008,2009,2010]
year_sum_1 = np.sum(dealer1, axis=1)
year_sum_2 = np.sum(dealer2, axis=1)
```

```python
# Draw bar chart for total products sold in each year
plt.bar(year, year_sum_1)
plt.title("Sales")
plt.xlabel("Year")
plt.ylabel("No.of Items")
plt.show()

plt.bar(year, year_sum_2,label="sale", color='red')
plt.xticks(year)
plt.title("Sales")
plt.xlabel("Year")
plt.ylabel("No.of Items")
plt.legend()
plt.show()
```

```python
# Pie char for average sales of each year
avg_sale_1 = np.average(dealer1, axis=1)
avg_sale_2 = np.average(dealer2, axis=1)

plt.pie(avg_sale_1, labels=year,autopct="%.1f%%")
plt.title('Sales for Dealer 1')
plt.show()

plt.pie(avg_sale_2, labels=year,autopct="%.1f%%", explode=(0,0,0,0,0,0.3,0,0,0,0), shadow=True)
plt.title('Sales for Dealer 2')
plt.show()
```
